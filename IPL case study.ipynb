```python
IPL["All_Players"]= IPL["Team1Players"]+IPL["Team2Players"]
def did_kohli_played(Player_list):
    return "V Kohli" in Player_list

IPL["Did_kohli_play"]=IPL["All_Players"].apply(did_kohli_played)
IPL[IPL[(IPL["City"]=="Delhi") & (IPL["Did_kohli_play"]==True)].drop_duplicates(subset=["City","Did_kohli_play"],keep="last")
```
