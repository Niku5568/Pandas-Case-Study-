```python
# 13: find average of each markes using apply method

def marks(x):
    return x//2

df1["Marks"].apply(marks)
```

```python
# 14: Add new colum of average marks of each student.
df1["Half Marks"]=df1["Marks"].apply(marks)
df1
```

```python
# 15: Add new colum of average marks of each student without floating values using lambda function.

df1["Avg Marks"]=df1["Marks"].apply(lambda x:x//2)
df1
```

```python
# 16: find len of each name and add new column using apply
df1["Name Length"]=df1["Name"].apply(len)
df1
```

```python
# 17 : ecode the gender as male = 1 and female = 0 using map function.

df1["encoded Gender"]=df1["Gender"].map({"Male":1,"Female":0})
df1
```

```python
# 18: drop any column
df1.drop("Half Marks",axis=1,inplace=True)
df1
```
